```python
plt.figure(figsize=(20, 6))
sns.scatterplot(data=df, x='Restaurant_longitude', y='Restaurant_latitude', hue='Delivery_person_Age', palette='viridis')
plt.title('Relationship Between Longitude and Latitude with the Delivery Person\'s Age')
plt.xlabel('Restaurant Longitude')
plt.ylabel('Restaurant Latitude')
plt.legend(title='Age', loc='upper left', bbox_to_anchor=(1, 1), frameon=False, fontsize=10)
plt.show()
```

```python
df['TARGET'] = pd.to_numeric(df['TARGET'], errors='coerce')

df.dropna(subset=['TARGET'], inplace=True)

plt.figure(figsize=(10, 6))
plt.hist(df['TARGET'], bins=30, color='blue', edgecolor='black')
plt.title('Delivery Time (TARGET) Distribution')
plt.xlabel('TARGET')
plt.ylabel('Frequency')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
