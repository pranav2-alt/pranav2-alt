```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/mall.csv")
```

```python
df.head(3)
```

```python
df = df.drop(columns = ['CustomerID' , 'Genre'])
```

```python
df.head(3)
```

```python
x= df.iloc[: , [0,1]].values
```

```python
from sklearn.cluster import KMeans
```

```python
import matplotlib.pyplot as plt
```

```python
a = []

for i in range(1,11):
  b = KMeans(n_clusters = i , init = 'k-means++' , random_state = 42)
  b.fit(x)
  a.append(b.inertia_)

plt.plot(range(1,11), a)

plt.title('The Elbow Method Graph')
plt.xlabel('Number of Clusters(k)')
plt.ylabel('wcs_list')
plt.show()
```

```python
b = KMeans(n_clusters=4, init='k-means++', random_state=42)
y_predict = b.fit_predict(x)
```

from the above plot, we can see the elbow pointis at 4. So the number of clusters here will be 4.

```python
plt.scatter(x[y_predict == 0, 0], x[y_predict == 0, 1], c='blue', label='Cluster 1')
plt.scatter(x[y_predict == 1, 0], x[y_predict == 1, 1], c='green', label='Cluster 2')
plt.scatter(x[y_predict == 2, 0], x[y_predict == 2, 1], c='orange', label='Cluster 3')
plt.scatter(x[y_predict == 3, 0], x[y_predict == 3, 1], c='purple', label='Cluster 4')

plt.scatter(b.cluster_centers_[:, 0], b.cluster_centers_[:, 1],
            s=300, c='yellow', marker='x', label='Centroids')

plt.xlabel('Cluster of customers')
plt.ylabel('Annual Income (k$)')
plt.title('Spending Score (1-100)')
plt.legend()
plt.show()
```
