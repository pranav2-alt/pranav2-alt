```python
import pandas as pd
import numpy as np

# Q.21 Add a new column to a DataFrame
df = pd.DataFrame({'Name': ['Alice', 'Bob'], 'Age': [25, 30]})
df['Country'] = 'USA'
print(df)

# Q.22 Remove a column from a DataFrame
df = df.drop(columns=['Country'])
print(df)

# Q.23 Filter rows based on multiple conditions
filtered = df[(df['Age'] > 25) & (df['Name'] == 'Bob')]
print(filtered)

# Q.24 Create a DataFrame from a NumPy array
arr = np.array([[1, 2], [3, 4]])
df = pd.DataFrame(arr, columns=['A', 'B'])
print(df)

# Q.25 Convert a column's data type
df['A'] = df['A'].astype(float)
print(df.dtypes)

# Q.26 Create a histogram of a column
df['A'].plot(kind='hist')

# Q.27 Export a DataFrame to a CSV file
df.to_csv('output.csv', index=False)

# Q.28 Create a DataFrame from a dictionary of lists
data = {'A': [1, 2], 'B': [3, 4]}
df = pd.DataFrame(data)
print(df)

# Q.29 Find the correlation between columns
correlation = df.corr()
print(correlation)

# Q.30 Convert a DataFrame to a NumPy array
arr = df.to_numpy()
print(arr)

# Q.31 Find the average salary for each department in a company
data = {'Department': ['HR', 'IT', 'HR', 'IT', 'Finance'], 'Salary': [40000, 50000, 45000, 55000, 60000]}
df = pd.DataFrame(data)
result = df.groupby('Department')['Salary'].mean()
print(result)

# Q.32 Count the number of employees in each department
result = df.groupby('Department')['Salary'].count()
print(result)

# Q.33 Calculate the total salary paid in each department
result = df.groupby('Department')['Salary'].sum()
print(result)

# Q.34 Find the mean and maximum salary for each department
result = df.groupby('Department')['Salary'].agg(['mean', 'max'])
print(result)

# Q.35 Find the total salary for each combination of department and job role
data = {'Department': ['HR', 'IT', 'HR', 'IT', 'Finance'], 'Role': ['Manager', 'Developer', 'Assistant', 'Developer', 'Analyst'], 'Salary': [40000, 50000, 45000, 55000, 60000]}
df = pd.DataFrame(data)
result = df.groupby(['Department', 'Role'])['Salary'].sum()
print(result)

# Q.36 Retrieve the first employee record from each department
result = df.groupby('Department').first()
print(result)

# Q.37 Find departments where the total salary is greater than 90,000
result = df.groupby('Department')['Salary'].sum()
filtered_result = result[result > 90000]
print(filtered_result)

# Q.38 Add a column that shows the cumulative salary in each department
df['Cumulative_Salary'] = df.groupby('Department')['Salary'].cumsum()
print(df)

# Q.39 Normalize salaries within each department by subtracting the group mean
df['Normalized_Salary'] = df.groupby('Department')['Salary'].transform(lambda x: x - x.mean())
print(df)

# Q.40 Find out how many employees are there in each department
result = df.groupby('Department').size()
print(result)
```
