```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(x=df['explicit'])
```

```python
sns.countplot(x='explicit', data=df)
plt.title('Count of Explicit vs Non-Explicit Podcasts')
plt.xlabel('Explicit Content')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(df.index, df['total_episodes'], color='blue', alpha=0.7)
plt.title("Scatter Plot of Total Episodes")
plt.xlabel("Index")
plt.ylabel("Total Episodes")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='media_type', y='total_episodes', data=df, palette="viridis", estimator='mean', ci=None)
plt.title("Average Episodes by Media Type", fontsize=16, fontweight='bold', color='darkgreen')
plt.xlabel("Media Type", fontsize=12)
plt.ylabel("Average Total Episodes", fontsize=12)
plt.show()
```

```python
df['media_type'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(8, 8), cmap='Set3', startangle=90)
plt.title("Distribution of Media Types", fontsize=16, fontweight='bold', color='darkblue')
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='media_type', data=df, palette='pastel')
plt.title("Count of Media Types", fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel("Media Type", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.show()
```
