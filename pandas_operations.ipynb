```python
import pandas as pd

# Q.1 Create a Pandas DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35], 'City': ['New York', 'Los Angeles', 'Chicago']}
df = pd.DataFrame(data)
print(df)

# Q.2 Read a CSV file into a Pandas DataFrame
df = pd.read_csv('data.csv')
print(df.head())

# Q.3 Check for missing values in a DataFrame
print(df.isnull().sum())

# Q.4 Drop rows with missing values
df_cleaned = df.dropna()
print(df_cleaned)

# Q.5 Fill missing values with a specific value
df_filled = df.fillna(0)
print(df_filled)

# Q.6 Filter rows based on a condition
filtered_df = df[df['Age'] > 30]
print(filtered_df)

# Q.7 Select a specific column in a DataFrame
ages = df['Age']
print(ages)

# Q.8 Rename columns in a DataFrame
df.rename(columns={'Name': 'Full Name', 'Age': 'Years'}, inplace=True)
print(df)

# Q.9 Sort a DataFrame by a specific column
sorted_df = df.sort_values(by='Age')
print(sorted_df)

# Q.10 Group data and calculate aggregate statistics
grouped = df.groupby('City').mean()
print(grouped)

# Q.11 Reset the index of a DataFrame
df.reset_index(drop=True, inplace=True)
print(df)

# Q.12 Merge two DataFrames
df1 = pd.DataFrame({'ID': [1, 2], 'Name': ['sam', 'raj']})
df2 = pd.DataFrame({'ID': [1, 2], 'Age': [25, 30]})
merged_df = pd.merge(df1, df2, on='ID')
print(merged_df)

# Q.13 Concatenate two DataFrames
df1 = pd.DataFrame({'A': [1, 2]})
df2 = pd.DataFrame({'A': [3, 4]})
concatenated = pd.concat([df1, df2])
print(concatenated)

# Q.14 Pivot a DataFrame
df = pd.DataFrame({'Name': ['Alice', 'Bob'], 'Year': [2020, 2021], 'Score': [85, 90]})
pivoted = df.pivot(index='Name', columns='Year', values='Score')
print(pivoted)

# Q.15 Melt a DataFrame
melted = pd.melt(df, id_vars=['Name'], value_vars=['Year', 'Score'])
print(melted)

# Q.16 Calculate the rolling mean of a column
df['Rolling Mean'] = df['Age'].rolling(window=2).mean()
print(df)

# Q.17 Find the unique values in a column
unique_cities = df['City'].unique()
print(unique_cities)

# Q.18 Count the occurrences of unique values in a column
value_counts = df['City'].value_counts()
print(value_counts)

# Q.19 Apply a custom function to a column
df['Age Squared'] = df['Age'].apply(lambda x: x**2)
print(df)

# Q.20 Drop duplicate rows in a DataFrame
df = df.drop_duplicates()
print(df)
```
